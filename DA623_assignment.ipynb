### IMPORTS

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.stats import multivariate_t # for 3rd distribution
```

### Generating synthetic data

#### Gaussian Distribution

```python
mean1 = np.array([0.0, 0.0])
cov1 = np.array([[10, 4],[4, 13]])
d1 = np.random.multivariate_normal(mean1, cov1, size=100)
d1_x = np.transpose(d1)[0]
d1_y = np.transpose(d1)[1]
```

#### Exponential Distribution

```python
lamda = 0.5
d2 = np.random.exponential(scale = lamda, size=(100,2))
d2_x = np.transpose(d2)[0]
d2_y = np.transpose(d2)[1]
```

#### Student-t Distribution

```python
mean3 = np.array([0.0, 0.0])
cov3 = np.array([[5.0, 2.0],[2.0, 6.2]])
d3 = multivariate_t.rvs(mean3, cov3, df=3, size=100)
d3_x = np.transpose(d3)[0]
d3_y = np.transpose(d3)[1]
```

### Principal Component Analysis for the generated data

#### Data1

```python
pca1 = PCA()
pca1.fit(d1)
eig1 = pca1.components_ # principal components
```

#### Data2

```python
pca2 = PCA()
pca2.fit(d2)
eig2 = pca2.components_ # principal components
```

#### Data3

```python
pca3 = PCA()
pca3.fit(d3)
eig3 = pca3.components_ # principal components
```

### Plots

```python
# plot for first data
plt.figure(1)
plt.title("Gaussian Distribution")
o1 = np.array([np.mean(d1_x), np.mean(d1_y)])
plt.quiver(o1[0], o1[1], eig1[0][0], eig1[0][1], scale=3, color='r', label='PC 1')
plt.quiver(o1[0], o1[1], eig1[1][0], eig1[1][1], scale=5, color='b', label='PC 2')
plt.scatter(d1_x, d1_y, color='k', label='data')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.xlabel("x")
plt.ylabel('y')
plt.legend()
plt.grid()
```

```python
# plot for 2nd data
plt.figure(2)
plt.title("Exponential Distribution")
o2 = np.array([np.mean(d2_x), np.mean(d2_y)])
plt.quiver(o2[0], o2[1], eig2[0][0], eig2[0][1], scale=3, color='r', label='PC 1')
plt.quiver(o2[0], o2[1], eig2[1][0], eig2[1][1], scale=5, color='b', label='PC 2')
plt.scatter(d2_x, d2_y, color='k', label='data')
plt.xlim(-0.5, 2)
plt.ylim(-0.5, 2)
plt.xlabel("x")
plt.ylabel('y')
plt.legend()
plt.grid()
```

```python
# plot for 3rd data
plt.figure(3)
plt.title("Student-t Distribution")
o3 = np.array([np.mean(d3_x), np.mean(d3_y)])
plt.quiver(o3[0], o3[1], -eig3[0][0], -eig3[0][1], scale=3, color='r', label='PC 1')
plt.quiver(o3[0], o3[1], eig3[1][0], eig3[1][1], scale=5, color='b', label='PC 2')
plt.scatter(d3_x, d3_y, color='k', label='data')
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.xlabel("x")
plt.ylabel('y')
plt.legend()
plt.grid()
```
